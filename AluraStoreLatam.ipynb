### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import pandas as pd

# Agregar columna que identifique la tienda
tienda["Tienda"] = "Tienda 1"
tienda2["Tienda"] = "Tienda 2"
tienda3["Tienda"] = "Tienda 3"
tienda4["Tienda"] = "Tienda 4"

# Unir todos los DataFrames
df_tiendas = pd.concat([tienda, tienda2, tienda3, tienda4])

df_tiendas.head()
```

```python
facturacion_tienda1 = tienda["Precio"].sum()
facturacion_tienda2 = tienda2["Precio"].sum()
facturacion_tienda3 = tienda3["Precio"].sum()
facturacion_tienda4 = tienda4["Precio"].sum()

print("Facturación Tienda 1:", facturacion_tienda1)
print("Facturación Tienda 2:", facturacion_tienda2)
print("Facturación Tienda 3:", facturacion_tienda3)
print("Facturación Tienda 4:", facturacion_tienda4)
```

##* Análisis de ingresos por tienda

```python
ingresos_por_tienda = df_tiendas.groupby("Tienda")["Precio"].sum()

ingresos_por_tienda.head()
```

```python
import matplotlib.pyplot as plt

tiendas = ["Tienda 1","Tienda 2","Tienda 3","Tienda 4"]

facturacion = [
facturacion_tienda1,
facturacion_tienda2,
facturacion_tienda3,
facturacion_tienda4
]

plt.bar(tiendas,facturacion)

plt.title("Ingresos por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos")

plt.show()
```

# 2. Ventas por categoría

```python
categoria1 = tienda["Categoría del Producto"].value_counts()
categoria2 = tienda2["Categoría del Producto"].value_counts()
categoria3 = tienda3["Categoría del Producto"].value_counts()
categoria4 = tienda4["Categoría del Producto"].value_counts()

print("Tienda 1")
print(categoria1)

print("Tienda 2")
print(categoria2)

print("Tienda 3")
print(categoria3)

print("Tienda 4")
print(categoria4)
```

```python
categoria1.plot(kind="bar")

plt.title("Ventas por categoría - Tienda 1")
plt.xlabel("Categoría")
plt.ylabel("Cantidad de ventas")

plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificacion1 = tienda["Calificación"].mean()
calificacion2 = tienda2["Calificación"].mean()
calificacion3 = tienda3["Calificación"].mean()
calificacion4 = tienda4["Calificación"].mean()

print("Promedio Tienda 1:",calificacion1)
print("Promedio Tienda 2:",calificacion2)
print("Promedio Tienda 3:",calificacion3)
print("Promedio Tienda 4:",calificacion4)
```

##* Promedio de reseñas por tienda

```python
reseñas_por_tienda = df_tiendas.groupby("Tienda")["Calificación"].mean()

reseñas_por_tienda.head()
```

```python
calificaciones = [calificacion1,calificacion2,calificacion3,calificacion4]

plt.bar(tiendas,calificaciones)

plt.title("Calificación promedio por tienda")
plt.xlabel("Tienda")
plt.ylabel("Calificación")

plt.show()
```

# 4. Productos más y menos vendidos

```python
productos_tienda1 = tienda["Producto"].value_counts()

print("Producto más vendido:")
print(productos_tienda1.head(1))

print("Producto menos vendido:")
print(productos_tienda1.tail(1))
```

##* Categorías más vendidas

```python
categorias_vendidas = df_tiendas.groupby("Categoría del Producto")["Producto"].count()

categorias_vendidas.sort_values(ascending=False)
```

##* Productos más vendidos

```python
productos_mas_vendidos = df_tiendas["Producto"].value_counts().head(10)

productos_mas_vendidos.head()
```

```python
productos_tienda1.head(10).plot(kind="bar")

plt.title("Top 10 productos más vendidos")
plt.xlabel("Producto")
plt.ylabel("Cantidad")

plt.show()
```

##* Ventas por Categorías de productos

```python
categoria1.plot(kind="pie",autopct="%1.1f%%")

plt.title("Distribución de ventas por categoría - Tienda 1")

plt.ylabel("")

plt.show()
```

# 5. Envío promedio por tienda

```python
envio1 = tienda["Costo de envío"].mean()
envio2 = tienda2["Costo de envío"].mean()
envio3 = tienda3["Costo de envío"].mean()
envio4 = tienda4["Costo de envío"].mean()

print("Envío promedio Tienda 1:",envio1)
print("Envío promedio Tienda 2:",envio2)
print("Envío promedio Tienda 3:",envio3)
print("Envío promedio Tienda 4:",envio4)
```

##* Costo promedio de envío

```python
envio_promedio = df_tiendas.groupby("Tienda")["Costo de envío"].mean()

envio_promedio.head()
```

```python
envios = [envio1,envio2,envio3,envio4]

plt.bar(tiendas,envios)

plt.title("Costo promedio de envío por tienda")
plt.xlabel("Tienda")
plt.ylabel("Costo de envío")

plt.show()
```
